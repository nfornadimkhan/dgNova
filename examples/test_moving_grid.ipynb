```python
import sys
import os

# Add the parent directory to Python path so we can import dgNova
sys.path.append(os.path.dirname(os.getcwd()))

# Now we can import DIALLEL
from dgNova.field_designs import UNREP
# 1) Import or paste the UNREP class definition (provided above) into your script or notebook.

# 2) Create an instance of the UNREP class with your data.
#    Here, we'll assume you have a DataFrame named `df` with columns:
#    'Row', 'Column', 'Genotype', 'Plot', and 'Yield'.
#    (You can also pass a CSV file path or a NumPy array.)

unrep_obj = UNREP(
    data='test_data/UNREP_Triticale_Unreplicated_Trial_Moving_Grids.csv',  # or path to CSV or a NumPy 2D array
    response='Yield',        # the response variable column
    row='Row',               # the row position column
    column='Column',         # the column position column
    genotype='Genotype',     # genotype column (optional but needed for replicate-based variance)
    plot='Plot',             # plot ID column
    design='moving_grid'     # the analysis design
)

# 3) Run the analysis:
results = unrep_obj.analyze()
```
