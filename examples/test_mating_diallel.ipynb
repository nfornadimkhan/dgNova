```python
import pandas as pd
import numpy as np
from dgNova.mating_designs import DIALLEL

# Read the parental cross data
data = pd.read_csv('test_data/parental-heterosis-data.csv')

# Create DIALLEL instance with method=1 (includes parents and F1s)
diallel = DIALLEL(data=data, 
                  parents=7,
                  method=1, 
                  response='yield',
                  parent1_col='p1',
                  parent2_col='p2')

# Run the complete analysis
results = diallel.analyze(silent=True)

# Calculate heterosis
het_results = diallel.calculate_heterosis()

# Show heterosis visualization
diallel.plot_heterosis()
```
